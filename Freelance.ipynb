## Dataset Overview

### Source of Data
The dataset used in this analysis is sourced from **Kaggle**, a popular platform for data science and machine learning projects. Kaggle hosts a wide range of datasets, and this specific dataset is focused on freelance job market data, providing key insights into the dynamics of global freelancing.

### Key Columns and Their Meanings
- **Freelancer_ID**: A unique identifier for each freelancer.
- **Job_Category**: The type of freelance work (e.g., Software Development, Design, Writing).
- **Platform**: The online platform where the job was posted (e.g., Upwork, Fiverr).
- **Experience_Level**: The freelancer's experience category (e.g., Entry, Mid-level, Expert).
- **Client_Region**: The geographical region of the client (e.g., North America, Europe, Asia).
- **Payment_Method**: The method of payment used for the job (e.g., PayPal, Bank Transfer).
- **Job_Completed**: A binary indicator (1/0) showing whether the job was successfully completed.
- **Earnings_USD**: The total earnings for the completed job in USD.
- **Hourly_Rate**: The hourly rate charged by the freelancer for the job.
- **Job_Success_Rate**: A rating (0–100%) indicating the freelancer's success rate on the platform.
- **Client_Rating**: The rating given by the client (on a scale of 1 to 5).
- **Job_Duration_Days**: The number of days taken to complete the job.
- **Project_Type**: The type of project (e.g., one-time task, ongoing contract).
- **Rehire_Rate**: The percentage of clients who have rehired the freelancer for subsequent projects.
- **Marketing_Spend**: The amount the freelancer spent on marketing to acquire clients.

# ****Machine Learning****

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV

# Load your dataset (replace this with your actual data loading method)
# For example: df = pd.read_csv("freelancer_data.csv")
df = pd.read_csv('freelancer_earnings_bd.csv')  # Replace with actual dataset path

# 1. Data Preprocessing
# Drop irrelevant columns (like Freelancer_ID)
df = df.drop(columns=['Freelancer_ID'])

# 2. Create a binary target variable for "highly rated" freelancers
df['Highly_Rated'] = (df['Client_Rating'] > 4).astype(int)  # Assuming highly rated if Client_Rating > 4

# Drop the target column from features for training
features = df.drop(columns=['Highly_Rated', 'Client_Rating'])

# 3. Encode categorical variables
label_encoders = {}
categorical_columns = ['Job_Category', 'Platform', 'Experience_Level', 'Client_Region', 
                       'Payment_Method', 'Project_Type']

# Apply Label Encoding for categorical columns
for col in categorical_columns:
    le = LabelEncoder()
    features[col] = le.fit_transform(features[col])
    label_encoders[col] = le

# 4. Scale numerical features
numerical_columns = ['Job_Completed', 'Earnings_USD', 'Hourly_Rate', 'Job_Success_Rate',
                     'Job_Duration_Days', 'Rehire_Rate', 'Marketing_Spend']

scaler = StandardScaler()
features[numerical_columns] = scaler.fit_transform(features[numerical_columns])

# 5. Train-Test Split
X = features
y = df['Highly_Rated']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6. Model Training using RandomForestClassifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# 7. Evaluate the model
y_pred = clf.predict(X_test)

# Display classification report and accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# 8. Hyperparameter Tuning (Optional)
# You can fine-tune your model with GridSearchCV to improve accuracy
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters found: ", grid_search.best_params_)

# Train with the best parameters
best_clf = grid_search.best_estimator_
y_pred_best = best_clf.predict(X_test)

print("Accuracy after tuning:", accuracy_score(y_test, y_pred_best))
print("Classification Report after tuning:\n", classification_report(y_test, y_pred_best))
```

**Interpretation:**
- The model predicts if a freelancer is **"highly rated"** (1) or **"not highly rated"** (0) based on factors like experience, earnings, job success rate, and client ratings.
 - The **accuracy** is 50%, meaning it correctly predicts 50% of the time.
 - **Precision (0.51)**: When the model predicts a freelancer is highly rated, it's correct 51% of the time.
 - **Recall (0.44)**: Out of all highly rated freelancers, it correctly identifies 44% of them.
 - **F1-score (0.47)**: This score balances precision and recall, showing the model's overall performance.

 **Conclusion:**
 - The model is not highly accurate, with room for improvement in identifying highly rated freelancers.
 - It provides a basic starting point for understanding factors that influence freelancer success but needs enhancement for more reliable predictions.
 - The results can help organizations identify trends and disparities but require further adjustments for actionable insights.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV

# Load dataset
df = pd.read_csv('freelancer_earnings_bd.csv')  # Replace with actual dataset path

# Data Preprocessing
df = df.drop(columns=['Freelancer_ID'])
df['Highly_Rated'] = (df['Client_Rating'] > 4).astype(int)  # Binary target: Highly Rated (1) or Not (0)
features = df.drop(columns=['Highly_Rated', 'Client_Rating'])

# Encode categorical variables
label_encoders = {}
categorical_columns = ['Job_Category', 'Platform', 'Experience_Level', 'Client_Region', 
                       'Payment_Method', 'Project_Type']
for col in categorical_columns:
    le = LabelEncoder()
    features[col] = le.fit_transform(features[col])
    label_encoders[col] = le

# Scale numerical features
numerical_columns = ['Job_Completed', 'Earnings_USD', 'Hourly_Rate', 'Job_Success_Rate',
                     'Job_Duration_Days', 'Rehire_Rate', 'Marketing_Spend']
scaler = StandardScaler()
features[numerical_columns] = scaler.fit_transform(features[numerical_columns])

# Train-Test Split
X = features
y = df['Highly_Rated']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train KNN Classifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# Evaluate KNN Model
y_pred_knn = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("KNN Classification Report:\n", classification_report(y_test, y_pred_knn))

# Hyperparameter Tuning (Optional)
param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # p=1 for Manhattan, p=2 for Euclidean
}
grid_search_knn = GridSearchCV(estimator=knn, param_grid=param_grid_knn, cv=5)
grid_search_knn.fit(X_train, y_train)

print("Best parameters for KNN: ", grid_search_knn.best_params_)

# Train with the best parameters
best_knn = grid_search_knn.best_estimator_
y_pred_best_knn = best_knn.predict(X_test)

print("KNN Accuracy after tuning:", accuracy_score(y_test, y_pred_best_knn))
print("KNN Classification Report after tuning:\n", classification_report(y_test, y_pred_best_knn))
```
